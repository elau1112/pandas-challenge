## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
data3Ng=pd.DataFrame(data1[["SN","Gender"]].value_counts())
data3Ng.head()
data3Ng.to_csv('data3Ng.csv',index=True, header=True)
data3 = pd.read_csv("data3Ng.csv")

data3Gen=data3['Gender'].value_counts() 
data3GenN=data3['Gender'].value_counts(normalize=True)*100
data3Gen=data3Gen.reset_index()
data3GenN=data3GenN.reset_index()
data3Gen.columns=[' ', 'Total Count']
data3GenN.columns=[' ', 'Percentages of Players']
merge_d3=pd.merge(data3Gen, data3GenN, on=' ')
merge_d3["Percentages of Players"]=merge_d3["Percentages of Players"].map("{:.2f}%".format)
merge_d3.sort_values(["Percentages of Players"], ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
d9=data1[['Gender','Price']]
d9g=d9.groupby(['Gender'])
d9g.head()
d9count=d9.groupby(['Gender']).count()
d9count=d9count.rename(columns={"Price":"Purchase Count"})

d9ave=d9.groupby(['Gender']).mean()
d9ave=d9ave.rename(columns={"Price":"Average Purchase Price"})

d9total=d9.groupby(['Gender']).sum()
d9total=d9total.rename(columns={"Price":"Total Purchase Value"})

merge_d9a=pd.merge(d9count, d9ave, on=['Gender'])
merge_d9b=pd.merge(merge_d9a,d9total, on=['Gender'])
merge_d3=merge_d3.rename(columns={" ":"Gender"})
merge_d9c=pd.merge(merge_d9b, merge_d3, on=['Gender'])
merge_d9c["Avg Total Purchase per Person"]=merge_d9c["Total Purchase Value"] / merge_d9c["Total Count"]
del merge_d9c['Total Count']
del merge_d9c['Percentages of Players']

merge_d9c["Average Purchase Price"]=merge_d9c["Average Purchase Price"].map("${:,.2f}".format)
merge_d9c["Total Purchase Value"]=merge_d9c["Total Purchase Value"].map("${:,.2f}".format)
merge_d9c["Avg Total Purchase per Person"]=merge_d9c["Avg Total Purchase per Person"].map("${:,.2f}".format)
merge_d9c.sort_values("Gender")

merge_d9c
```

```python
#The following is another method that I have tried, but without the groupby used
pcount=data1['Gender'].value_counts()
pcount=pcount.reset_index()
pcount.columns=['Gender', 'Purchase Count']

data1_female_price=data1.loc[data1["Gender"] == "Female","Price"].mean()
data1_male_price=data1.loc[data1["Gender"] == "Male","Price"].mean()
data1_other_price=data1.loc[data1["Gender"] == "Other / Non-Disclosed","Price"].mean()

data1_female_pur=data1.loc[data1["Gender"] == "Female","Price"].sum()
data1_male_pur=data1.loc[data1["Gender"] == "Male","Price"].sum()
data1_other_pur=data1.loc[data1["Gender"] == "Other / Non-Disclosed","Price"].sum()

atppp_female=data1_female_pur / merge_d3["Total Count"][1]
atppp_male=data1_male_pur / merge_d3["Total Count"][0]
atppp_other=data1_other_pur / merge_d3["Total Count"][2]

d4a=pd.DataFrame({"Gender":["Female", "Male","Other / Non-Disclosed"],
                  "Average Purchase Price":[data1_female_price, data1_male_price, data1_other_price],
                  "Total Purchase Value":[data1_female_pur, data1_male_pur, data1_other_pur],
                  "Avg Total Purchase per Person":[atppp_female, atppp_male, atppp_other]
                 })
merge_d4a=pd.merge(pcount, d4a, on='Gender')

merge_d4a["Average Purchase Price"]=merge_d4a["Average Purchase Price"].map("${:,.2f}".format)
merge_d4a["Total Purchase Value"]=merge_d4a["Total Purchase Value"].map("${:,.2f}".format)
merge_d4a["Avg Total Purchase per Person"]=merge_d4a["Avg Total Purchase per Person"].map("${:,.2f}".format)
merge_d4a.sort_values("Gender")
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
data5Na=pd.DataFrame(data1[["SN","Age"]].value_counts())
data5Na.head()
data5Na.to_csv('data5Na.csv',index=True, header=True)
data5 = pd.read_csv("data5Na.csv")

bins = [0, 9, 14, 19, 24, 29, 34, 39, data5["Age"].max()+1]
agerg=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
data5["Age Ranges"]=pd.cut(data5["Age"], bins, labels=agerg, include_lowest=True)
del data5["SN"]
del data5["0"]
data5=data5.rename(columns={"Age":"Total Count"})
data5=data5.groupby("Age Ranges").count()

data5.to_csv('data5a.csv', index=True, header=True)
data5a=pd.read_csv("data5a.csv")
data5a["Percentage of Players"]=data5a["Total Count"]/data5a["Total Count"].sum()*100
data5a["Percentage of Players"]=data5a["Percentage of Players"].map("{:.2f}%".format)
data5a
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
data6=data1[["Age","Price"]]
data6.to_csv('data6ag.csv',index=True, header=True)
data6a = pd.read_csv("data6ag.csv")

data6a["Age Ranges"]=pd.cut(data6a["Age"], bins, labels=agerg, include_lowest=True)
del data6a["Unnamed: 0"]


data6b=data6a.groupby("Age Ranges").count()

data6b=data6b.rename(columns={"Age":"Purchase Count"})
data6bave=data6a.groupby("Age Ranges").mean()
data6bave=data6bave.rename(columns={"Price":"Average Purchase Price"})
data6btotal=data6a.groupby("Age Ranges").sum()
data6btotal=data6btotal.rename(columns={"Price":"Total Purchase Value"})

del data6b["Price"]
del data6btotal["Age"]
del data6bave["Age"]

merge_d6c=pd.merge(data6b, data6bave, on=['Age Ranges'])
merge_d6d=pd.merge(merge_d6c,data6btotal, on=['Age Ranges'])
merge_d6d
```

```python
merge_d6e=pd.merge(merge_d6d, data5a, on=['Age Ranges'])
merge_d6e["Avg Total Purchase per Person"]=merge_d6e["Total Purchase Value"] / merge_d6e["Total Count"]
del merge_d6e['Total Count']
del merge_d6e['Percentage of Players']
merge_d6e["Average Purchase Price"]=merge_d6e["Average Purchase Price"].map("${:,.2f}".format)
merge_d6e["Total Purchase Value"]=merge_d6e["Total Purchase Value"].map("${:,.2f}".format)
merge_d6e["Avg Total Purchase per Person"]=merge_d6e["Avg Total Purchase per Person"].map("${:,.2f}".format)
merge_d6e
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
data7=data1[["SN","Price"]]

data7bcount=data7.groupby(['SN']).count()
data7bcount=data7bcount.rename(columns={"Price":"Purchase Count"})

data7b=data7.groupby(['SN']).sum()
data7b=data7b.rename(columns={"Price":"Total Purchase Value"})


data7bave=data7.groupby(['SN']).mean()
data7bave=data7bave.rename(columns={"Price":"Average Purchase Price"})

merge_d7=pd.merge(data7bcount, data7bave, on='SN')
merge_d7b=pd.merge(merge_d7, data7b, on='SN')

merge_d7b=merge_d7b.sort_values("Total Purchase Value", ascending=False)
merge_d7b["Average Purchase Price"]=merge_d7b["Average Purchase Price"].map("${:,.2f}".format)
merge_d7b["Total Purchase Value"]=merge_d7b["Total Purchase Value"].map("${:,.2f}".format)
merge_d7b.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
data8=data1[["Item ID","Item Name","Price"]]
data8g=data8.groupby(['Item ID', 'Item Name'])
data8g.head()

data8bcount=data8.groupby(['Item ID', 'Item Name']).count()
data8bcount=data8bcount.rename(columns={"Price":"Purchase Count"})

data8bip=data8.groupby(['Item ID', 'Item Name']).mean()
data8bip=data8bip.rename(columns={"Price":"Item Price"})

data8bsum=data8.groupby(['Item ID', 'Item Name']).sum()
data8bsum=data8bsum.rename(columns={"Price":"Total Purchase Value"})

merge_83a=pd.merge(data8bip, data8bsum, on=['Item ID', 'Item Name'])
merge_83b=pd.merge(data8bcount, merge_83a, on=['Item ID', 'Item Name'])

merge_83c=merge_83b.sort_values("Purchase Count", ascending=False)
merge_83c["Item Price"]=merge_83c["Item Price"].map("${:,.2f}".format)
merge_83c["Total Purchase Value"]=merge_83c["Total Purchase Value"].map("${:,.2f}".format)
merge_83c.head()
```
